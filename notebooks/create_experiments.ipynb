```python
from langgraph.graph import StateGraph,MessagesState,START,END
from langchain_core.messages import SystemMessage, HumanMessage, BaseMessage
from get_llm import get_llm
```

```python
#get th llm

llm =  get_llm(model = "gemini-2.5-flash-lite")

#create node
def ask_llm(state:MessagesState):
    role = [SystemMessage(content="You are a Popular web seach application")]
    result = llm.invoke(
        role + state["messages"]
    )
    return {"messages":result}

```

```python
graph = StateGraph(MessagesState)

#addnodes
graph.add_node("Ask_llm",ask_llm)

#add edges
graph.set_entry_point("Ask_llm")
graph.set_finish_point("Ask_llm")

compiled_graph = graph.compile()
compiled_graph
```

```python
response = compiled_graph.invoke({
            "messages":[HumanMessage(content = "Why Gachibowli is named Gachibowli?")]
            }
            )
response["messages"][-1].pretty_print()
```
